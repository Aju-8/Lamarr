```python
#Web Scraping Coding Project

import requests
import pandas as pd
from bs4 import BeautifulSoup
import json
```

```python
new_url='https://spicyip.com'
a=[]
p=[]
for page in range(1,6):
  url2=new_url+'/page/'+str(page)
  response=requests.get(new_url)
  soup1= BeautifulSoup(response.content,'html.parser')
  new=soup1.find_all('header',class_='entry-header')

  for i in new:
    new1=i.find('a').text.strip()
    a.append(new1)
  para=soup1.find_all('div',class_='entry-content clear')

  for j in para:
    new2=j.find('p').text.strip()
    p.append(new2)
```

```python
data_list = []
page_data={'Title':a,'Content':p}
data_list.append(page_data)
```

```python
with open('pages_data.json', 'w', encoding='utf-8') as json_file:
    json.dump(data_list, json_file, ensure_ascii=False, indent=4)
```
